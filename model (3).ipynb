```python
# Define a function to plot distribution
def plot_distribution(distribution, title):
    # Sort values
    sorted_distribution = distribution.sort_values(ascending=False)  
    random_colors = [(random.uniform(0.6, 1), random.uniform(0.6, 1), random.uniform(0.6, 1)) for _ in sorted_distribution]
    
    # Set figure size
    plt.figure(figsize=(14, 6))  
    
    # Plot pie chart
    plt.subplot(1, 2, 1)
    plt.pie(sorted_distribution, labels=sorted_distribution.index, autopct='%1.1f%%', colors=random_colors, startangle=100, textprops={'fontsize': 12})
    plt.title(title, fontsize=20)
    plt.axis('equal')  # Set aspect ratio to keep the pie chart circular

    # Display distribution values on the right side
    plt.subplot(1, 2, 2)
    plt.text(0, 1, '\n'.join(f'{index}: {value}' for index, value in sorted_distribution.items()), fontsize=12, va='top')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
# Define cities
cities = data['City'].unique()

# Define age groups
age_groups = ['Early Working', 'Prime Working', 'Mature Working', 'Elderly','Children']

# Define marital statuses and colors
marital_colors = {'Never Married': {'Men': 'blue', 'Women': 'cyan'},
                  'Married': {'Men': 'green', 'Women': 'magenta'}}

# Set up subplots for each city
fig, axs = plt.subplots(1, len(cities), figsize=(6*len(cities), 8), sharey=True)

# Loop through each city
for i, city in enumerate(cities):
    # Filter data for the current city
    city_data = data[data['City'] == city]
    
    # Initialize counts for each age group
    total_counts = {age_group: 0 for age_group in age_groups}
    
    # Loop through each marital status
    for marital_status, marital_color in marital_colors.items():
        # Loop through each gender
        for gender in ['Men', 'Women']:
            # Filter data for the current marital status and gender
            filtered_data = city_data[(city_data['Marital Status'] == marital_status) & (city_data['Gender'] == gender)]
            # Count occurrences of each age group for the current marital status and gender
            age_counts = filtered_data['Age'].value_counts().reindex(age_groups, fill_value=0)
            # Update total counts for each age group
            total_counts = {age_group: total_counts[age_group] + count for age_group, count in age_counts.items()}
            # Create positions for bars
            positions = np.arange(len(age_groups))
            # Plot data
            axs[i].bar(positions, age_counts, color=marital_color[gender], width=0.4, label=f'{marital_status} {gender}')
    
    # Set labels and title
    axs[i].set_xticks(positions)
    axs[i].set_xticklabels(age_groups)
    axs[i].set_title(city)
    axs[i].legend(loc='lower right')
    axs[i].grid(axis='y')

# Set common labels
fig.text(0.5, 0.0, 'Age Group', ha='center', fontsize=15)
fig.text(0.0, 0.5, 'Population Count', va='center', rotation='vertical', fontsize=15)
plt.tight_layout()
plt.show()
```

```python
# Define a function to extract percentage invested
def extract_percentage(percentage_string):
    if isinstance(percentage_string, str):
        if 'Above' in percentage_string:  # If it's in the 'Above xx%' format
            return float(percentage_string.split(' ')[1].replace('%', ''))
        elif 'Upto' in percentage_string:  # If it's in the 'Upto xx%' format
            return float(percentage_string.split(' ')[1].replace('%', ''))
        elif 'Don\'t Want' in percentage_string:  # If it's a placeholder value
            return None
        else:  # If it's in the 'xx% to yy%' format
            return (int(percentage_string.split(' to ')[0].replace('%', '')) + int(percentage_string.split(' to ')[1].replace('%', ''))) / 2
    else:
        return None

# Apply the function to extract numerical values
data['Percentage_Invested'] = data['Percentage of Investment'].apply(extract_percentage)

# Calculate average percentage invested per household for each combination of age and number of households
percentage_stats = data.groupby(['Age', 'Number of investors in family'])['Percentage_Invested'].mean().reset_index()

# Pivot the data for better visualization
pivot_percentage = percentage_stats.pivot(index='Age', columns='Number of investors in family', values='Percentage_Invested')

# Plotting
plt.figure(figsize=(10, 8))
heatmap = plt.pcolor(pivot_percentage, cmap='coolwarm', vmin=0, vmax=pivot_percentage.max().max())  # Plot heatmap
plt.colorbar(heatmap)  # Add color bar
plt.title('Percentage Invested by Age Group and Number of Investors in family')
plt.xlabel('Number of Households')
plt.ylabel('Age Group')
plt.xticks(np.arange(len(pivot_percentage.columns)) + 0.5, pivot_percentage.columns, rotation=45)  # Adjust x ticks
plt.yticks(np.arange(len(pivot_percentage.index)) + 0.5, pivot_percentage.index)  # Adjust y ticks
plt.gca().invert_yaxis()  # Invert y-axis to match with the heatmap
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {
    'Awareness_Source': np.random.choice(['Financial Advisor', 'Internet Research', 'Family/Friends', 'Seminars'], 100),
    'Knowledge_Level': np.random.choice(['Low', 'Moderate', 'High'], 100),
    'Risk_Level': np.random.choice(['Low', 'Moderate', 'High'], 100),
    'Reasons_for_Investment': np.random.choice(['Retirement', 'Wealth Accumulation', 'Education', 'Emergency Fund'], 100)
}
df = pd.DataFrame(data)

# Calculate PMI for categorical variables
def pmi(x, y):
    p_x = x.value_counts(normalize=True)
    p_y = y.value_counts(normalize=True)
    p_xy = pd.crosstab(x, y).div(len(x))
    return np.log(p_xy.div(np.outer(p_x, p_y)))

# Plot PMI heatmaps
fig, axs = plt.subplots(2, 2, figsize=(16, 12))

pmi_matrix1 = pmi(df['Awareness_Source'], df['Knowledge_Level'])
sns.heatmap(pmi_matrix1, annot=True, cmap='coolwarm', fmt=".2f", ax=axs[0, 0])
axs[0, 0].set_title('Awareness Source vs Knowledge Level')

pmi_matrix2 = pmi(df['Awareness_Source'], df['Reasons_for_Investment'])
sns.heatmap(pmi_matrix2, annot=True, cmap='coolwarm', fmt=".2f", ax=axs[0, 1])
axs[0, 1].set_title('Awareness Source vs Reasons for Investment')

pmi_matrix3 = pmi(df['Reasons_for_Investment'], df['Knowledge_Level'])
sns.heatmap(pmi_matrix3, annot=True, cmap='coolwarm', fmt=".2f", ax=axs[1, 0])
axs[1, 0].set_title('Reasons for Investment vs Knowledge Level')

pmi_matrix4 = pmi(df['Risk_Level'], df['Knowledge_Level'])
sns.heatmap(pmi_matrix4, annot=True, cmap='coolwarm', fmt=".2f", ax=axs[1, 1])
axs[1, 1].set_title('Risk Level vs Knowledge Level')

plt.tight_layout()
plt.show()
```
